```python
import openreview

# API V2
client = openreview.api.OpenReviewClient(
    baseurl='https://api2.openreview.net',
    username='cowolff@uni-osnabrueck.de',
    password='Avatar30.11'
)
```

```python
print(openreview.tools.get_all_venues(client))
```

```python
submissions = client.get_all_notes(content={'venueid':'ICML.cc/2024/Conference'} )
```

```python
# API 2
results = []
for accepted_submission in submissions:
    current = {}
    current['title'] = accepted_submission.content['title']['value']
    author_profiles = openreview.tools.get_profiles(client, accepted_submission.content['authorids']['value'])
    current['authors'] = []
    for author_profile in author_profiles:
        current_author = {}
        current_author['name'] = author_profile.get_preferred_name(pretty=True)
        try:
            current_author['affiliation'] = author_profile.content.get('history', [{}])[0]['institution']['name']
        except:
            current_author['affiliation'] = 'N/A'
        current['authors'].append(current_author)
    results.append(current)

print(results)
```

```python
# store in json
import json
with open('data/ICML2024.json', 'w') as f:
    json.dump(results, f, indent=4)
```
